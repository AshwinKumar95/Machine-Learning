```python
#CHARACTERS
```

```python
def character_count(words):
    char_count = []
    for i in range(0, len(words)):
        char_count.append(len(words[i]))
    final_char_count = sum(char_count)
    return(final_char_count)
```

```python
print('The number of characters in the speech by President  Roosevelt in 1941 is:', character_count(roosevelt.split()))
print('The number of characters in the speech by President  Kennedy in 1961 is:', character_count(kennedy.split()))
print('The number of characters in the speech by President Nixon in 1973 is:', character_count(nixon.split()))
```

```python
#REMOVAL OF STOP WORDS
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
def stopwords_remove(speech):
    speech = speech.lower()
    words = []
    for word in speech.split():
        if word not in set(stopwords.words('english')):
            words.append(word)
        speech = ' '.join(words)
    return(speech)    
```

```python
stopwords_roosevelt = stopwords_remove(roosevelt)
stopwords_kennedy = stopwords_remove(kennedy)
stopwords_nixon = stopwords_remove(nixon)
```

```python
stopwords_count_roosevelt = len(roosevelt.lower().split()) - len(stopwords_roosevelt.split())
stopwords_count_kennedy = len(kennedy.lower().split()) - len(stopwords_kennedy.split())
stopwords_count_nixon = len(nixon.lower().split()) - len(stopwords_nixon.split())
```

```python
print('The number of stop words present in the speech by President Roosevelt in 1941 is:', stopwords_count_roosevelt)
print('The number of stop words present in the speech by President Kennedy in 1961 is:', stopwords_count_kennedy)
print('The number of stop words present in the speech by President Nixon in 1973 is:', stopwords_count_nixon)
```

```python
# SPECIAL CHARACTER REMOVAL

from collections import Counter
```

```python
filtered_roosevelt = re.sub('[^\w\s]', '', stopwords_roosevelt)
filtered_kennedy = re.sub('[^\w\s]', '', stopwords_kennedy)
filtered_nixon = re.sub('[^\w\s]', '', stopwords_nixon)
```

```python
counts_roosevelt = Counter(filtered_roosevelt.split())
counts_roosevelt.most_common(10)
```

```python
counts_kennedy = Counter(filtered_kennedy.split())
counts_kennedy.most_common(10)
```

```python
counts_nixon = Counter(filtered_nixon.split())
counts_nixon.most_common(10)
```
